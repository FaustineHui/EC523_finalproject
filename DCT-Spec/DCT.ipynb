```python
import numpy as np
from scipy import fftpack
from PIL import Image
from copy import deepcopy
import os
```

```python
train_dirs = ['progan_train', 'progan_val']
train_categories = ['car', 'cat', 'horse', 'person', 'sofa']
labels = ['0_real', '1_fake']

for train_dir in train_dirs:
    for train_category in train_categories:
        for label in labels:
            direc = 'C:/Users/fuzha/Downloads/' + train_dir + '/' + train_category + '/' + label
            images = os.listdir(direc)

            for image in images:
                img = Image.open(direc + '/' + image)
                img = np.array(img)

                im = deepcopy(img[16:240,16:240,:])
                im = im.astype(np.float32)
                im = im/255.0
                for i in range(3):
                    img = im[:,:,i]
                    dct_img = fftpack.dct(img, type=2, norm="ortho", axis=0)
                    dct_img = fftpack.dct(dct_img, type=2, norm="ortho", axis=1)
                    dct_img = np.log(np.abs(dct_img)+1e-3)
                    dct_min = np.percentile(dct_img,5)
                    dct_max = np.percentile(dct_img,95)
                    dct_img = (dct_img - dct_min)/(dct_max - dct_min)
                    dct_img = (dct_img-0.5)*2
                    dct_img[dct_img<-1] = -1
                    dct_img[dct_img>1] = 1
                    im[:,:,i] = dct_img

                im = np.transpose(im, (2,0,1))

                img = Image.fromarray(np.uint8(im[0]))
                img.save('C:/Users/fuzha/Downloads/DCT/' + train_dir + '/' + train_category + '/' + label + '/' + image)
```
